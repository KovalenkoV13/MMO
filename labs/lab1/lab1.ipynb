```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("players_data_light-2024_2025.csv")
plt.rcParams["figure.figsize"] = (12, 6)
```

Распределение возрастов игроков

```python
plt.figure()
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Распределение возрастов игроков')
plt.xlabel('Возраст')
plt.ylabel('Количество игроков')
plt.show()
```

Связь возраста и игрового времени

```python
plt.figure()
sns.scatterplot(x=df['Age'], y=df['Min'])
plt.title('Возраст vs Игровое время')
plt.xlabel('Возраст')
plt.ylabel('Минуты на поле')
plt.show()
```

Голы и передачи по позициям

```python
df_grouped = df.groupby('Pos')[['Gls', 'Ast']].mean().sort_values(by='Gls', ascending=False)
df_grouped.plot(kind='bar', stacked=True)
plt.title('Среднее количество голов и передач по позициям')
plt.xlabel('Позиция')
plt.ylabel('Среднее значение')
plt.show()
```

Распределение команд по лигам

```python
plt.figure()
df['Comp'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Распределение игроков по лигам')
plt.ylabel('')
plt.show()
```

Зависимость числа матчей от результативности

```python
plt.figure()
sns.lineplot(x=df['MP'], y=df['G+A'])
plt.title('Число матчей vs Голы+Передачи')
plt.xlabel('Матчи')
plt.ylabel('Голы + Передачи')
plt.show()
```

Выводы:
- Большинство игроков находится в возрасте 22-30 лет.
- В среднем, старшие игроки играют меньше минут.
- Нападающие забивают больше всего голов, полузащитники отдают больше передач.
- Английская Премьер-лига представлена наибольшим количеством игроков.
- Чем больше матчей играет игрок, тем выше его общая результативность.
